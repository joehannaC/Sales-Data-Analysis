## Identify top-performing products and services

since rating is in object format, we need to convert it into numeric format

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
print("Lowest rating:", df['rating'].min())
print("Highest rating:", df['rating'].max())
```

#### **Rating Scale Interpretation**

- Ratings were categorized into five levels ranging from very low to excellent to facilitate clearer interpretation of customer satisfaction.

    0.0 ‚Äì 1.0	Very Low / Poor

    1.1 ‚Äì 2.0	Low

    2.1 ‚Äì 3.0	Average

    3.1 ‚Äì 4.0	Good

    4.1 ‚Äì 5.0	Excellent



```python
def rating_label(r):
    if r <= 1.0:
        return "Very Low"
    elif r <= 2.0:
        return "Low"
    elif r <= 3.0:
        return "Average"
    elif r <= 4.0:
        return "Good"
    else:
        return "Excellent"

df['rating_level'] = df['rating'].apply(rating_label)
```

```python
df.head()
```

Remove comma in 'rating_count' column then convert it into integer format

```python
df['rating_count'] = (
    df['rating_count']
    .astype(str)
    .str.replace(',', '', regex=False)
)

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

print("Minimum rating count:", df['rating_count'].min())
print("Maximum rating count:", df['rating_count'].max())
```

```python
avg_rating_count = df['rating_count'].mean()
print("Average rating count:", avg_rating_count)
```

```python
rating_threshold = 4.1
rating_count_threshold = 18270.564505119455

top_products = df[
    (df['rating'] >= rating_threshold) &
    (df['rating_count'] > rating_count_threshold)
]

top_10 = top_products.sort_values(by='rating_count', ascending=False).head(10)

top_10[['product_name', 'about_product', 'rating', 'rating_count', 'rating_level']]
```

```python
top_10['short_name'] = top_10['about_product'].str[:10] + '...'

x = np.arange(len(top_10))
width = 0.4

fig, ax1 = plt.subplots(figsize=(12,6))

bars_count = ax1.bar(x - width/2, top_10['rating_count'], width, color='skyblue', label='Rating Count')
ax1.set_xlabel('Product')
ax1.set_ylabel('Rating Count', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

for bar, count in zip(bars_count, top_10['rating_count']):
    ax1.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{int(count):,}',
             ha='center', va='bottom', fontsize=9)

ax2 = ax1.twinx()
bars_rating = ax2.bar(x + width/2, top_10['rating'], width, color='orange', label='Rating')
ax2.set_ylabel('Rating', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

for bar, rating in zip(bars_rating, top_10['rating']):
    ax2.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{rating:.1f}',
             ha='center', va='bottom', fontsize=9)

plt.xticks(x, top_10['short_name'], rotation=45, ha='right')
plt.title('Top 10 Performing Products: Rating Count & Rating')
fig.tight_layout()
plt.show()
```

To identify the top-performing products, two criteria were applied: an average product rating of at least 4.1 and a rating count higher than the overall average (18,270). Products meeting both criteria indicate high customer satisfaction and strong engagement. Figure above presents the top 10 performing products based on these thresholds. The dual-axis bar chart displays Rating Count on the left axis and Rating (1‚Äì5) on the right axis.

From the figure, it is evident that the products with the highest number of reviews also maintain high average ratings, reflecting both popularity and quality. This analysis provides actionable insights into which products are most successful, combining both quantitative engagement (number of ratings) and qualitative feedback (rating). Such products can be considered benchmark items for understanding customer preferences and guiding future product strategies.

## Analyze monthly and quarterly sales trends

Since there is no 'date' or 'sale_date' column, we create a new column 'sale_date' in which we will randomly assign dates starting from 01-01-2025.

```python
np.random.seed(42)

df['sale_date'] = pd.to_datetime('2025-01-01') + pd.to_timedelta(
    np.random.randint(0, 365, size=len(df)), unit='D'
)

df['month'] = df['sale_date'].dt.to_period('M')
df['quarter'] = df['sale_date'].dt.to_period('Q')
```

```python
df.head()
```

Clean the 'discounted_price' column

```python
df['discounted_price'] = df['discounted_price'].str.replace('‚Çπ', '', regex=False).str.replace(',', '', regex=False)
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')

print(df['discounted_price'].head())
```

Compute the revenue by multiplying the values in 'discounted_price' column to the values in 'rating_count'

Why 'rating_count'? 

- Since there's no record for the quantity, we will get the quantity by using the total counts of the rating

```python
df['revenue'] = df['discounted_price'] * df['rating_count']
```

```python
df.head()
```

Compute for monthly and quarterly revenue

```python
df['sale_date'] = pd.to_datetime(df['sale_date'])

df['month'] = df['sale_date'].dt.to_period('M')
df['quarter'] = df['sale_date'].dt.to_period('Q')

monthly_revenue = df.groupby('month')['revenue'].sum()
quarterly_revenue = df.groupby('quarter')['revenue'].sum()
```

```python
df.head()
```

Plot the monthly revenue

```python
plt.figure(figsize=(12,6))
monthly_revenue.plot(marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue (‚Çπ)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The monthly revenue for 2025 shows fluctuations throughout the year. The highest monthly revenue was observed in June 2025, reaching approximately ‚Çπ9.15 billion, indicating peak sales activity. Other months with strong performance include February (‚Çπ8.10B), April (‚Çπ7.59B), and September (‚Çπ7.36B). On the other hand, November (‚Çπ3.08B) and March (‚Çπ3.17B) recorded the lowest revenues, suggesting potential off-peak periods for sales. This trend may reflect seasonal demand, promotional campaigns, or product launches affecting customer purchases.

Plot the quarterly revenue

```python
plt.figure(figsize=(10,6))
quarterly_revenue.plot(kind='bar', color='skyblue')
plt.title('Quarterly Revenue Trend')
plt.xlabel('Quarter')
plt.ylabel('Revenue (‚Çπ)')
plt.tight_layout()
plt.show()
```

Looking at the quarterly performance, Q2 2025 generated the highest revenue at approximately ‚Çπ22.98 billion, driven mainly by strong sales in April and June. Q1 2025 followed with ‚Çπ18.12 billion, while Q3 2025 recorded ‚Çπ18.23 billion. The lowest quarterly revenue was in Q4 2025, totaling ‚Çπ12.58 billion, likely due to lower sales activity in November and December. This quarterly breakdown highlights periods of high demand, which can be used for strategic planning, marketing campaigns, and inventory management.
